```python
import warnings
import re
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import make_pipeline

BALANCED_DATA_CSV = 'data/balanced.csv'
ORIGINAL_DATA_CSV = 'data/original.csv'
TEST_CSV = 'data/test.csv'

warnings.filterwarnings('ignore')

class TextClassifier:
    def __init__(self):
        self.labelEncoder = LabelEncoder()
        self.pipeline = make_pipeline(
            TfidfVectorizer(),
            MultinomialNB()
        )
        self.gs = None
        
    def _transform_text(self, text_array):
        def to_lower(text):
            return text.lower()
        
        def remove_spaces(text):
            return " ".join(text.split()).strip()
        
        def sanatize(text):
            return re.sub('[.:;/$@&*\'"]', '', text) 
        
        return [to_lower(remove_spaces(sanatize(text))) for text in text_array]
    
    def train(self, csv_file, test_size=0.3):
        df = pd.read_csv(csv_file, header=0)
        x = self._transform_text(df[df.columns[0]])
        y = self.labelEncoder.fit_transform(df[df.columns[1]])

        xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=test_size)

        self.gs = GridSearchCV(self.pipeline, param_grid={
                'multinomialnb__fit_prior': [True, False],  
                'multinomialnb__force_alpha': [True, False],
                'multinomialnb__alpha': [i/10 for i in range(1, 30)]
            }, return_train_score=True)
        
        self.gs.fit(xtrain, ytrain)

        print("Training complete.")
        print("Best parameters:", self.gs.best_params_)
        print("Score on validation set:", self.gs.score(xtest, ytest))

        return self.gs.best_params_, self.gs.score(xtest, ytest)

    def test(self, csv_file):
        if not self.gs:
            raise ValueError("Model has not been trained yet.")

        df = pd.read_csv(csv_file, header=0)
        x_test = self._transform_text(df[df.columns[0]])
        y_test = df[df.columns[1]]

        y_test_pred = self.gs.predict(x_test)

        print("Testing complete.")
        print("Best parameters:", self.gs.best_params_)
        print("Score on test set:", self.gs.score(x_test, self.labelEncoder.transform(y_test)))

        table = pd.DataFrame(
            zip(x_test, [self.labelEncoder.classes_[result] for result in y_test_pred], y_test),
            columns=["Input", "Prediction", "Expected"]
        )
        print("Results: ")

        return table

```

```python
classifier = TextClassifier()
classifier.train(BALANCED_DATA_CSV, test_size=0.3)
classifier.test(TEST_CSV)
```

```python
classifier = TextClassifier()
classifier.train(ORIGINAL_DATA_CSV)
classifier.test(TEST_CSV)
```
